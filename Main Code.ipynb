### Replacing outliers with median

```python
def handle_outliers(x, column):
    Q1 = x[column].quantile(0.25)
    Q3 = x[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    x[column] = x[column].apply(lambda l: x[column].median() if l < lower_bound or l > upper_bound else l)
    return x
for column in x.columns:
  x = handle_outliers(x,column)
```

```python
x.plot(kind='box', figsize=(17,17), layout=(4,4), sharex=False, subplots=True)
```

```python
for column in ['accelerations', 'percentage_of_time_with_abnormal_long_term_variability', 'mean_value_of_long_term_variability', 'histogram_mode', 'histogram_variance']:
  x = handle_outliers(x,column)
```

```python
x.plot(kind='box',figsize=(17,17),layout=(4,4),sharex=False,subplots=True)
```

```python
x = handle_outliers(x,'accelerations')
x = handle_outliers(x,'histogram_mode')
x = handle_outliers(x,'prolongued_decelerations')
x = handle_outliers(x,'histogram_variance')
x = handle_outliers(x,'histogram_median')
x = handle_outliers(x,'mean_value_of_long_term_variability')
x = handle_outliers(x,'percentage_of_time_with_abnormal_long_term_variability')
```
